# Importing and visualizing molecules

In RDKit, molecules are typically imported from text-based representations such as SMILES strings. These are converted into RDKit Mol objects, which store the molecular structure and allow further analysis, visualization, and property calculations.

<div class="alert alert-success"><b> üîß Import the RDKit module.</b>

```python
# Import the necessary RDKit module

from rdkit import Chem # RDKit‚Äôs core chemistry module.
```

Import Draw: Visualizing Molecules

```python
from rdkit.Chem import Draw # Render molecules in 2D
```

To calculate molecular properties, we use the Descriptors module in RDKit. These properties are computed directly from the molecular structure: once a molecule is converted into a Mol object, RDKit analyzes its atoms and bonds and applies built-in chemical rules to calculate numerical descriptors such as molecular weight, logP, and hydrogen bond counts. These values allow us to quantitatively describe and compare molecules.

**Import Descriptors:**

```python
from rdkit.Chem import Descriptors # Compute molecular descriptors (numeric properties) 

# Including molecular weight, cLogP, H-bond donors/acceptors, topological polar surface area (TPSA), and more.
```

<div class="alert alert-success"><b> üîß Canonical SMILES</b>

Canonical SMILES is a standardized, unique text representation of a molecule. While a molecule can have multiple valid SMILES strings, the canonical form ensures one consistent representation for comparison and database searches. In RDKit, you can generate and print the canonical SMILES to reliably identify a molecule.

**1. Provide a SMILES string**

Note: SMILES codes usually omit hydrogens; they are implicit, so the Smiles string "CCO" represents ethanol even though the hydrogen on the oxygen is not written.

```python
# Create a molecule object from a SMILES string

smiles_input = 'CCO'  # smiles_input=‚Äòthe chemical formula‚Äô
```

**2. Parse into an RDKit Mol**</br>
Parsing a molecule means converting its textual representation, such as a SMILES string, into an RDKit Mol object. This object stores the molecule‚Äôs atoms, bonds, and structure, allowing RDKit to visualize it, analyze it, or calculate molecular properties.

```python
# Transfer the smiles to Mol

mol = Chem.MolFromSmiles(smiles_input)  # mol= Chem.MolFromSmiles()
```

**3. Generate a canonical SMILES**

```python
# Transfer the Mol to Smiles 
canonical_smiles = Chem.MolToSmiles(mol) # canonical_smiles= Chem.MolToSmiles()
```

**4. Print the result**

```python
print(f"Canonical SMILES: {canonical_smiles}")

# f means an f-string.
```

**Loop over atoms bonds and print info**<br>
mol.GetBonds() returns all bonds in the molecule, and the for loop goes through each one.
bond.GetBeginAtomIdx() and bond.GetEndAtomIdx() give the indices of the two atoms connected by the bond.
bond.GetBondType() shows the type of the bond (e.g., SINGLE, DOUBLE, TRIPLE, AROMATIC).

```python
# Loop through bonds in the molecule

print("\nBonds in the molecule:") # \n is used to move the text to a new line, commonly used for formatting output.

for bond in mol.GetBonds(): # mol.GetBonds() yields bond objects;
    print(f"Bond: {bond.GetBeginAtomIdx()} - {bond.GetEndAtomIdx()}, Type: {bond.GetBondType()}")
    # GetBeginAtomIdx() and GetEndAtomIdx() give endpoint atom indices;
    # GetBondType() reports the bond type (e.g. SINGLE, DOUBLE, TRIPLE, AROMATIC, etc.).
```

## Calculating molecular properties ##

```python
# Define molecules and their SMILES
molecules = {
    "Acetone": "CC(=O)C",
    "Benzene": "c1ccccc1",
    "Aspirin": "CC(=O)OC1=CC=CC=C1C(=O)O"
}
```

```python
# Use RDKit's Chem.MolFromSmiles function to convert the SMILES string into an RDKit Mol object.

for name, smiles_input in molecules.items():
    mol = Chem.MolFromSmiles(smiles_input)
```

**Use RDKit's Descriptors module to calculate the following molecular properties**

```python
# Use RDKit's Descriptors module to calculate the following molecular properties.

for name, smiles_input in molecules.items():

    mw = Descriptors.MolWt(mol)  # molar weight
    logp = Descriptors.MolLogP(mol) # logp
    donors = Descriptors.NumHDonors(mol) # donors
    acceptors = Descriptors.NumHAcceptors(mol) # acceptors
```

**Print the results**

```python
# Print the calculated molecular weight, LogP, hydrogen bond donors, and hydrogen bond acceptors.

for name, smiles_input in molecules.items():
    print(f"\n==============================")
    print(f" Molecule: {name}")
    print(f"Molecular weight: {mw:.2f}") # ‚Äú.2f‚Äù means format as fixed-point with two decimal places.
    print(f"LogP: {logp:.2f}")
    print(f"H-bond donors: {donors}")
    print(f"H-bond acceptors: {acceptors}")
```

# Lipinski Rule-of-Five 

**Lipinski‚Äôs Rule of Five (Ro5):**  
This rule is used to evaluate the drug-likeness of a molecule, i.e., whether a compound is likely to be orally active in humans. According to Lipinski‚Äôs Ro5, a molecule is more likely to be orally bioavailable if it meets the following criteria:

- **Molecular Weight (MW):** Less than 500 daltons.  
- **log P (Octanol-water partition coefficient):** Does not exceed 5.  
- **Hydrogen Bond Donors (HBD):** Does not exceed 5.  
- **Hydrogen Bond Acceptors (HBA):** Does not exceed 10.  
- **Rule of violations:** Molecules violating more than one of these rules may have poor absorption or permeation.  
> **Note:** It is called the ‚ÄúRule of Five‚Äù because all the numerical thresholds are multiples of five, not because there are five descriptors.


<div class="alert alert-success"><b> üîß Calculate the the number of Lipinski Rule-of-Five violations for each molecule.</b>

To predict **Lipinski's Ro5 violations** , you can make use of the 4 molecular descriptors:
1. **MW** (Molecular weight),  A molecular mass is less than **500 daltons**.
2. **log P** (Octanol-water partition coefficient), log P does not exceed **5**.
3. **HBD** (Hydrogen bond donors), HBD does not exceed **5**.
4. **HBA** (Hydrogen bond acceptors), HBA does not exceed **10**.

**Import matplotlib.pyplot to plot the data**

```python
import matplotlib.pyplot as plt
```

**Import pandas to analyse the results**

```python
import pandas as pd
```

**The moleuclar data set**

```python
# Input the smiles strings for defining the moleclue set

molecules = {
    "Aspirin": "CC(=O)OC1=CC=CC=C1C(=O)O",
    "Ibuprofen": "CC(C)CC1=CC=C(C=C1)C(C)C(=O)O",
    "Caffeine": "Cn1cnc2n(C)c(=O)n(C)c(=O)c12",
    "Penicillin": "CC1(C)SCC(N1C(=O)C2=CC=CC=C2)C(=O)O",
    "Cholesterol": "CC(C)CCCC(C)C1CCC2C1(CCC3C2CCC4=C3C=CC=C4)C"
}
```

**Define the Ro5 violation counter** <br>
This code block (function) counts how many of Lipinski‚Äôs Ro5 thresholds a molecule violates.
It calculates MW, logP, HBD, and HBA, adding 1 to the counter for each value exceeding its threshold.
The block returns the total number of violations (0‚Äì4), showing how drug-like the molecule is.

```python
# Define the Ro5 violation counter using def function.
# Compute MW, LogP, HBD, HBA and compare to thresholds.
# Add 1 per exceeded threshold (range 0‚Äì4).

def ro5_violations(mol):
    mw = Descriptors.MolWt(mol)
    logp = Descriptors.MolLogP(mol)
    hbd = Descriptors.NumHDonors(mol)
    hba = Descriptors.NumHAcceptors(mol)
    violations = 0  # the violations is countered from 0
    if mw > 500:    # the violations is plused 1 if the molar weight exceed threshold (500)
        violations += 1
    if logp > 5:    # the violations is plused 1 if the logp exceed threshold (5)
        violations += 1
    if hbd > 5:     # the violations is plused 1 if the hbd exceed threshold (5)
        violations += 1
    if hba > 10:    # the violations is plused 1 if the hba exceed threshold (10)
        violations += 1
    return violations  # return the vaule of violations
```

**Parse SMILES to RDkit Mol**

```python
mols = [Chem.MolFromSmiles(smi) for smi in molecules.values()]
```

**Build a simple table-like object** <br>
This code creates a pandas DataFrame with two columns: "Name" for molecule names and "Mol" for RDKit Mol objects.
The names come from the keys of the molecules dictionary‚Äîin Python, a dictionary stores data as key‚Äìvalue pairs, and the keys are the labels used to access each value (here, the molecule names).
The DataFrame makes it easy to compute properties, visualize molecules, and apply functions like Lipinski‚Äôs Ro5 to many molecules at once.

```python
# Create two columns‚Äîname and Mol‚Äîfor convenient batch computation and plotting.
import pandas as pd
df = pd.DataFrame({
    "Name": molecules.keys(),
    "Mol": mols
})
```

**Calculate the number of violations per molecule** <br>
This line applies the ro5_violations function to each molecule in the "Mol" column and creates a new column "ro5_violations" in the DataFrame. Each row will shows a number from 0 to 4, representing how many of Lipinski‚Äôs Ro5 thresholds that molecule violates.

```python
# Apply the counter to each Mol, producing a 0‚Äì4 violations column.

df["ro5_violations"] = df["Mol"].apply(ro5_violations)
```

**Visualize as a bar chart**

```python
# what elements for a bar chart needed,  x label, y label, title, limitation of y axis
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5)) # 1.figure size
plt.bar(df["Name"], df["ro5_violations"], color='darkgreen') # 2. bar of each moleclues
plt.xlabel("Molecule") # 3. x label
plt.ylabel("Number of Ro5 Violations") # 4. y label
plt.title("Lipinski‚Äôs Rule of Five Violations per Molecule") # 5. title of chart
plt.ylim(0, 4) # 6. limitation of y axis
plt.show()
```
